```python
plt.figure(figsize=(6,4))
sns.countplot(x="type", data=df)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="rating", data=df, order=df["rating"].value_counts().index)
plt.title("Content Ratings")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["listed_in"].value_counts().head(10).plot(kind="bar")
plt.title("Top Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["release_year"], bins=30)
plt.title("Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

df["Year Added"] = df["date_added"].dt.year

plt.figure(figsize=(10,5))
df["Year Added"].value_counts().sort_index().plot()
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["director"].value_counts().head(10).plot(kind="bar")
plt.title("Top Directors")
plt.show()
```

```python
actors = df["cast"].dropna().str.split(", ").explode()

plt.figure(figsize=(10,5))
actors.value_counts().head(10).plot(kind="bar")
plt.title("Top Actors")
plt.show()
```

```python
numeric = df.select_dtypes(include="number")

plt.figure(figsize=(8,5))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.show()
```
